```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, complete
import scipy
import matplotlib
import numpy as np
import pandas as pd
import sklearn.datasets as dts
```

```python
dataset = dts.load_wine()
data_wine = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
data_wine.head()
```

```python
points_linkage = data_wine.iloc[:,[0,1]].values
points_complete = data_wine.iloc[:,[0,4]].values
```

```python
linkage_hieararchical = linkage(points_linkage, method='ward')
complete_hieararchical = complete(points_complete)
```

```python
plt.figure(1, figsize=(25, 20), facecolor="white")
dendrogram(linkage_hieararchical)
plt.title('Dendrogram linkage')
plt.xlabel('Data Point')
plt.ylabel('Distance')

plt.figure(2, figsize=(25, 20), facecolor="white")
dendrogram(complete_hieararchical)
plt.title('Dendrogram complete')
plt.xlabel('Data Point')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aglo_linkage = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
aglo_complete = AgglomerativeClustering(n_clusters=4, affinity='euclidean')
```

```python
pr_linkage = aglo_linkage.fit_predict(points_linkage)
pr_complete = aglo_complete.fit_predict(points_complete)
```

```python
plt.figure(1, figsize=(10,10))
plt.scatter(points_linkage[pr_linkage==0, 0],points_linkage[pr_linkage==0, 1], s=25, c='red')
plt.scatter(points_linkage[pr_linkage==1, 0],points_linkage[pr_linkage==1, 1], s=25, c='green')
plt.scatter(points_linkage[pr_linkage==2, 0],points_linkage[pr_linkage==2, 1], s=25, c='blue')
plt.title('Linkage Aglomerative')
plt.ylabel('Malic Acid')
plt.xlabel('Alcohol')

plt.figure(2, figsize=(10,10))
plt.scatter(points_complete[pr_complete==0, 0],points_complete[pr_complete==0, 1], s=25, c='red')
plt.scatter(points_complete[pr_complete==1, 0],points_complete[pr_complete==1, 1], s=25, c='green')
plt.scatter(points_complete[pr_complete==2, 0],points_complete[pr_complete==2, 1], s=25, c='blue')
plt.scatter(points_complete[pr_complete==3, 0],points_complete[pr_complete==3, 1], s=25, c='purple')
plt.title('Linkage Aglomerative')
plt.ylabel('Magnesium')
plt.xlabel('Alcohol')

plt.show
```
