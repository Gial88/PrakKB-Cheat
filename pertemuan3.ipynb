<h1> #DATA VISUALIZATION</h1>

```python
#IMPORT PACKAGE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_categorical = pd.read_csv('House_Rent_Dataset.csv')
dataset_categorical.head(10)
```

```python
dataset_timeSeries = pd.read_csv('DailyDelhiClimateTest.csv')
dataset_timeSeries.head(10)
```

```python
plt.bar(dataset_categorical['City'], dataset_categorical['Rent'], color='aquamarine', 
        width=0.5, align='center')
plt.title("Perbandingan nilai sewa antar kota")
plt.xlabel("Nama Kota")
plt.ylabel("WC")
plt.show()
```

```python
date = [item for item in range(0, len(dataset_timeSeries['date']))]

plt.plot(date, dataset_timeSeries['humidity'], 'b', marker='o')
plt.plot(date,dataset_timeSeries['meantemp'], 'r')
plt.plot(date,dataset_timeSeries['wind_speed'], 'g', linestyle='dashed')
plt.legend(['Kelembapan', 'Rata - Rata Suhu', 'Kecepatan Angin'],loc='upper right')
plt.title('Data Kelembapan')
plt.xlabel('Waktu')
plt.ylabel('Kelembapan')
plt.show()
```

```python
dataset_categorical['City'].value_counts()
```

```python
labels = dataset_categorical['City'].value_counts().rename_axis("Kota").reset_index(name='Jumlah')
```

```python
labels.head()
```

```python
plt.figure(figsize=(5,5), dpi=1000)
plt.pie(dataset_categorical['City'].value_counts(), labels=labels['Kota'], 
        autopct='%.1f%%', startangle=90)
plt.show
```

```python
sns.scatterplot(data=dataset_categorical, x = 'Size', y='Rent')
sns.scatterplot(data=dataset_categorical, x = 'Bathroom', y='Rent')
plt.title("Pesebaran Data")
plt.show()
```

```python
sns.boxplot(x=dataset_categorical['Rent'], y=dataset_categorical['City'] )
plt.show()
```
