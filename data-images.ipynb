```python
import os
import shutil
import re
import random
```

```python
root_path = r'D:\ketukar\datasets'
source_path = os.path.join(root_path, 'data_gambar')
destiantion_path = os.path.join(root_path, 'data_gambar_used')
```

```python
folders = {}


for folder in os.listdir(source_path):
    folders[folder] = len(os.listdir(os.path.join(source_path, folder)))

folders
```

```python
if os.path.isdir(os.path.join(destiantion_path, 'training')):
    print(f"path {os.path.join(destiantion_path, 'training')} sudah ada")
else:
    os.makedirs(os.path.join(destiantion_path, 'training'))
    for folder in folders.keys():
            os.makedirs(os.path.join(destiantion_path, f'training/{folder}'))

if os.path.isdir(os.path.join(destiantion_path, 'validation')):
    print(f"path {os.path.join(destiantion_path, 'validation')} sudah ada")
else:
    os.makedirs(os.path.join(destiantion_path, 'validation'))
    for folder in folders.keys():
            os.makedirs(os.path.join(destiantion_path, f'validation/{folder}'))
```

```python
files_name = {
    "paper":[],
    "rock":[],
    "scissors":[],
}
files_train = {
    "paper":[],
    "rock":[],
    "scissors":[],
}
files_val = {
    "paper":[],
    "rock":[],
    "scissors":[],
}
```

```python
for type in files_name.keys():
    for file in os.listdir(os.path.join(source_path, type)):
        if os.path.getsize(os.path.join(source_path, f'{type}/{file}')):
            files_name[type].append(file)
        else:
            print(f'{file} rusak')
```

```python
files_name['scissors']
```

```python
split_point = {
    "paper":0,
    "rock":0,
    "scissors":0,
}

for key in split_point.keys():
    split_point[key] = int(len(files_name[key])*.2)

split_point

for key in files_train.keys():
    files_train[key].append(files_name[key][split_point[key]:])
    files_val[key].append(files_name[key][:split_point[key]])

len(files_train['paper'][0])
```

```python
for type in files_name.keys():
    for file in files_train[type][0]:
        shutil.copyfile(
            os.path.join(source_path, f'{type}/{file}'),
            os.path.join(destiantion_path, f'training/{type}/{file}'),
        )

for type in files_name.keys():
    for file in files_val[type][0]:
        shutil.copyfile(
            os.path.join(source_path, f'{type}/{file}'),
            os.path.join(destiantion_path, f'validation/{type}/{file}'),
        )
```
