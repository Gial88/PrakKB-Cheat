```python
arra1 = tf.constant([
    [1,2,3],
    [2,4,8],
    [1,3,5],
])
arra2= tf.constant([
    [2,3,5],
    [7,11,13],
    [17,23,29]
])
```

```python
print(arra1.shape)
print(arra2.shape)
arra_gabung = tf.concat([arra1, arra2], axis=0)
print(arra_gabung.shape)
print(arra_gabung)
```

```python
print(arra1.shape)
print(arra2.shape)
arra_stack = tf.stack([arra1, arra2], axis=1)
print(arra_stack.shape)
print(arra_stack)
```

```python
print(arra_stack.shape)
arra_unstack = tf.unstack(arra_stack, axis=1)
print(arra_unstack)
```

```python
arra_split = tf.split(arra_stack, num_or_size_splits=3)
print(arra_split)
```
