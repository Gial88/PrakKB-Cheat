```python
import tensorflow as tf
import numpy as np
```

```python
xdata = tf.constant([1,2,3,4,5,6])
ydata = tf.constant([2,4,6,8,10,12])
```

### Data Modeling Steps 
- buat arsiktektur model
- konfigurasi model
- latih model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])
model.summary() # unit*(input+1)
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
model.fit(
    x=xdata,
    y=ydata,
    epochs=500,
    steps_per_epoch=3
)
```
